```python
%matplotlib inline
import os, json, collections, statistics
from collections import Counter
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

FILE_PATH = "/home/jingbozhou/Project/CombDrug/runAddMut/9-plot/1-NumberofData/drugPath/"
```

```python
### 5. plot drug distribution
drug2path_df = pd.read_csv(os.path.join(FILE_PATH, "drug_pathway.csv"), index_col="Drug_name")

all_count_d = Counter(drug2path_df["Pathway"].values)
all_count_d = {k: v for k, v in sorted(all_count_d.items(), key=lambda item: item[1], reverse=True)}
label_list, count_list = [], []
for key, value in all_count_d.items():
    label_list.append(key)
    count_list.append(value)
    
ylocs = np.arange(len(count_list))

sns.set(style="ticks", font_scale=1.2, rc={'figure.figsize':(6, 6)})
fig, ax = plt.subplots()

ax.barh(label_list, count_list, align='center', height=1,
        color=list(sns.color_palette("RdBu", len(ylocs))))
    
for x, y in zip(count_list, ylocs):
    ax.text(x, y, x, va='center')
            
ax.invert_yaxis()

ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

#plt.savefig("./plotFigure/drug_pathway_dis.svg", format="svg", bbox_inches="tight", transparent=True)
```
